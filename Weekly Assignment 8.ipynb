## Read in CSV file

```python
import pandas as pd

denver_data = pd.read_csv('denver-1.nov.csv')  
denver_data.head()
```

## Group Data by Date

Find the total number of arrests, citations, and warnings issued each day in November. 
Drop the columns that we don't need: the raw row number and lat/long.

```python
stats_daily = denver_data.groupby(['date']).sum()
stats_daily = stats_daily.drop(['raw_row_number','lat', 'lng'], axis=1)
stats_daily
```

## Export the Data as a .json

With json, turn the dataframe into a json file using split to get the columns, data, and date (index) in the right format. Write it to a file that we can use in JS.

```python
import json
import pprint as pprint

result = stats_daily.to_json(orient="split")
parsed = json.loads(result)
pprint.pprint(json.dumps(parsed, indent=4))

with open('wa8.json', 'w') as f:
    f.write(result)
```
